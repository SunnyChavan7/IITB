```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# 1. Pie Chart using Matplotlib
fig, ax = plt.subplots()
x = [8, 9, 48, 21, 33, 242]
label = ['8 label','Pressured, for some other reason', 'Pressured to take, for education or work', 'Pressured to take for travel', 'Willingly, for others health', '242 label']
ax.pie(x, labels=label, autopct='%.1f%%')
ax.set_title('Quantification od Covid 19 Vaccine Coercion in India: A Survey Study')
plt.show()
```

```python
# 2. Vertical bar chart with plotly
import plotly.express as px
df = pd.DataFrame({'x2': ['A', 'B', 'C', 'D', 'E'], 'y2': [-1.8, 0.1, 0.2, 0.3, 4.8]})
fig = px.bar(df, x='x2', y='y2')
fig.show()
```

```python
# 3. Horizontal bar chart using matplotlib
y=['2023', '2022', '2021', '2020', '2019']
x=[19000, 17000, 18000, 15000, 15000]
plt.barh(y, x)
plt.ylabel("Year")
plt.xlabel("Under 45-deaths in first 43 weeks (England and Wales)")
plt.title("UK Data : Injection IS the pandemic for < 45")
plt.show()
```
